```python
import numpy as np
import pandas as pd
from math import radians, sin, cos, sqrt, atan2

# Load real estate and subway station data
real_estate_data = data_JS  # Replace with your real estate data file
subway_station_data = pd.read_csv('/content/drive/MyDrive/팀 프로젝트 /ppt 참고 코드(통계분석, 머신러닝,딥러닝)/서울지하철주소종합.csv', encoding = 'utf-8')  # Replace with your subway station data file

# Function to calculate Haversine distance
def haversine_distance(lat1, lon1, lat2, lon2):
    R = 6371000  # Radius of Earth in meters
    phi1 = radians(lat1)
    phi2 = radians(lat2)

    delta_phi = radians(lat2 - lat1)
    delta_lambda = radians(lon2 - lon1)

    a = sin(delta_phi / 2) * sin(delta_phi / 2) + cos(phi1) * cos(phi2) * sin(delta_lambda / 2) * sin(delta_lambda / 2)
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c  # Distance in meters
    return distance

# 컬럼명 변환
subway_station_data = subway_station_data.rename(columns = {'위도': 'latitude', '경도': 'longitude'})

# Calculate shortest distance for each real estate to each subway station
shortest_distances = []

for _, real_estate_row in unique_rows.iterrows():
    real_estate_lat = real_estate_row['latitude']
    real_estate_lon = real_estate_row['longitude']

    distances_to_subway_stations = []
    for _, subway_station_row in subway_station_data.iterrows():
        subway_station_lat = subway_station_row['latitude']
        subway_station_lon = subway_station_row['longitude']

        distance = haversine_distance(real_estate_lat, real_estate_lon, subway_station_lat, subway_station_lon)
        distances_to_subway_stations.append(distance)

    shortest_distance = min(distances_to_subway_stations)
    shortest_distances.append(shortest_distance)

# Add the shortest distances to subway stations to the real estate data
unique_rows['Shortest_Distance_to_Subway'] = shortest_distances

# Save the merged data to a new CSV file
unique_rows.to_csv('/content/drive/MyDrive/팀 프로젝트 /ppt 참고 코드(통계분석, 머신러닝,딥러닝)/data_JS.csv', index=False)

data_JS = unique_rows

data_JS
```
