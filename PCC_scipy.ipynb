## PCC

```python
# 데이터 타입 변경
df['Year'] = df['Year'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Sell_Count'] = df['Sell_Count'].astype(float)
df['JS_Count'] = df['JS_Count'].astype(float)
```

```python
result_df = pd.DataFrame(columns=['Column_Name', 'PCC', 'p-value'])

# JS_Price와 다른 열 간의 PCC 및 p-value 계산 및 저장
for column in df.columns:
    if column != 'JS_Price':
        if df[column].dtype in (int, float):
            correlation, p_value = pearsonr(df['JS_Price'], df[column])
            result_df = result_df.append({'Column_Name': column, 'PCC': correlation, 'p-value': p_value}, ignore_index=True)
```

```python
result_df.sort_values(by='PCC', ascending=False)
```
